```python
import sklearn.datasets
df = sklearn.datasets.load_iris(as_frame=True)['data']
df = df.iloc[:10, :10]
df
```

```python
from IPython.display import display, HTML

def make_pretty(styler):
    styler.set_caption("Weather Conditions")
    styler.background_gradient(axis="index", vmin=0, vmax=1, cmap="YlGnBu", subset=['petal width (cm)'])
    # styler.text_gradient(cmap="Reds")
    return styler

# def pretty_display_df_res(df):
#     display(HTML(
#         df
#         .style
#         .background_gradient(cmap='Reds')
#         .to_html()
#     ))
#     # .highlight_between(left=left, right=1.5, axis=1, props='color:white; background-color:purple;')

def pretty_display_df_res(df):
    h = df.style.pipe(make_pretty).to_html()
    # print(h)
    display(HTML(h))
    # return df
    # .highlight_between(left=left, right=1.5, axis=1, props='color:white; background-color:purple;')

pretty_display_df_res(df)
```
